```python
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
import pandas as pd
import time
from datetime import datetime
import requests
newslst = []
for pagenum in range(1,7):
    print(pagenum)
    url = 'https://www.euronews.com/search?query=gdpr&p={}'.format(pagenum)
    response = requests.get(url).text
    one_page = BeautifulSoup(response)
    resultlst = one_page.find_all("article")
    for result in resultlst:
        resultdata = result.find("a",{'class':"m-object__title__link "})
        title = resultdata["title"]
        url = 'euronews.com' + resultdata["href"]
        date = result.find("time", {"class":"m-object__date u-margin-top-2"}).text.split('\n')[1]
        newslst.append({'title':title, 'url':url,'date':date, 'page':pagenum})
newsdf = pd.DataFrame(newslst)
newsdf
```

```python
newsdf['contents'] = ''
for i in range(len(newsdf)):
    url = 'https://' + newsdf.loc[i,'url']
    response = requests.get(url).text
    one_page = BeautifulSoup(response)
    contentslst = one_page.find_all("p")
    neirong = ''
    for j in range(1,len(contentslst)):
        neirong += contentslst[j].text
        neirong += ' '
    neirong = neirong.replace("Share this article",'')
    neirong = neirong.replace("Learn more about",'')
    print(i)
    newsdf.loc[i,'contents'] = neirong
newsdf
```

```python
newsdf.to_excel("Euronews Crawling_0101.xlsx",index=False)
```
