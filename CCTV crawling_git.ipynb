```python
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
import pandas as pd
import time
from datetime import datetime
import requests
newslst = []#31
for i in range(1,31):
    print(i)
    url = 'https://search.cctv.com/search.php?qtext=%E4%B8%AA%E4%BA%BA%E4%BF%A1%E6%81%AF%E4%BF%9D%E6%8A%A4%E6%B3%95&sort=relevance&type=web&vtime=&datepid=1&channel=&page={}'.format(i)
    response = requests.get(url).text
    one_page = BeautifulSoup(response)
    resultlst = one_page.find_all("li")
    print(len(resultlst))
    for result in resultlst:
        try:
            title = result.find("a").text
            url = result.find('span')['lanmu1']
            date = result.find("span",{"class":"tim"}).text[5:15]
            newslst.append({'title':title,'url':url,'date':date})
        except KeyError:
            pass

newsdf = pd.DataFrame(newslst)
newsdf
```

```python
newsdf['content'] = ''
for i in range(300):
    print(i)
    nr = ''
    url = newsdf.loc[i,'url']
    response = requests.get(url)
    response.encoding = response.apparent_encoding
    response = response.text
    one_page = BeautifulSoup(response)
    contentlst = one_page.find_all('p')
    for content in contentlst:
        a = content.text
        if 'var' in a:
            nr+= ''
        else:
            nr += a
    nr = nr.replace("\u3000",'')
    nr = nr.replace("\n",'')
    nr = nr.replace("\xa0",'')
    nr = nr.replace("\r",'')
    nr = nr.replace("\t",'')
    nr = nr.replace("中央广播电视总台央视网版权所有",'') 
    newsdf.loc[i,'content'] = nr
newsdf
```

```python
newsdf.to_excel("CCTV Crawling_2.xlsx",index=False)
```
