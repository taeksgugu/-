```python
import pandas as pd
df = pd.read_excel("微博爬取内容55.xlsx")
df
```

```python
df.columns
```

```python
df = df[['发文时间', '文章内容']]
df
```

```python
def timecnt(x):
    a = x.split(' ')[1]
    if a == 'Dec':
        a = '12'
    elif a == 'Nov':
        a = '11'
    elif a == 'Oct':
        a = '10'
    elif a == 'Sep':
        a = '09'
    elif a == 'Aug':
        a = '08'
    elif a == 'Jul':
        a = '07'
    elif a == 'Jun':
        a = '06'
    elif a == 'May':
        a = '05'
    elif a == 'Apr':
        a = '04'
    elif a == 'Mar':
        a = '03'
    elif a == 'Feb':
        a = '02'
    elif a == 'Jan':
        a = '01'

    return x.split(' ')[-1] + '-' + a + '-' + x.split(' ')[2]
```

```python
df['Timestamp'] = df['发文时间'].apply(lambda x: timecnt(x))
df = df[['Timestamp', '文章内容']]
df.columns = ['Timestamp', 'Text']
df
```

```python
df.sort_values(by='Timestamp')
```

```python
totaldf = df.loc[(df['Timestamp']>='2020-10-01')&(df['Timestamp']<='2022-01-01')]
totaldf
```
