```python
import pandas as pd
df = pd.read_excel("Euronews Crawling_0101.xlsx")
df
```

```python
import datetime
df['date'] = df['date'].apply(lambda x: x.replace(" ",''))
df['date2'] = df['date'].apply(lambda x: x.replace('/','.'))
df
```

```python
dd = list(df.date3.unique())
dd.sort()
dd
```

```python
from datetime import datetime
df['date3'] = df['date2'].apply(lambda x: datetime.strftime(datetime.strptime(x, "%d.%m.%Y"),'%Y-%m-%d'))
df                               
```

```python
df2 = df.loc[df['date3']>='2018.05.01']
df2 = df2[['title', 'date3', 'contents']]
df2
```

```python
df3 = df2.loc[df2['contents'].str.contains("GDPR")]
df3
```

```python
df3.to_excel("Euronews 가공.xlsx", index=False)
```
