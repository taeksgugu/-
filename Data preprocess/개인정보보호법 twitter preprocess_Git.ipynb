### 가공_날짜

```python
import pandas as pd
df = pd.read_excel('개인정보보호법 twitter 가공.xlsx')
df
```

```python
df['Time'] = df['Timestamp'].apply(lambda x: str(x).split('T')[0])
df
```

```python
ddf = df.loc[df['Time'] >= '2020-02-01']
ddf
```

```python
ddf.to_excel('개인정보보호법 twitter 가공_날짜 가공.xlsx')
```
