```python
from selenium.webdriver import Chrome
from bs4 import BeautifulSoup
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
import pandas as pd
import time
from datetime import datetime
import requests
newslst = []
for i in range(1,106):
    print(i)
    url = 'https://news.sbs.co.kr/news/search/main.do?query=%EA%B0%9C%EC%9D%B8%EC%A0%95%EB%B3%B4%EB%B3%B4%ED%98%B8%EB%B2%95&pageIdx={}&searchOption=1&collection='.format(i)
    driver = Chrome('C:\\Users\\onrik\\chromedriver.exe')
    driver.get(url)
    response = driver.page_source
    one_page = BeautifulSoup(response)
    resultlst = one_page.find_all("a",{'class':'psil_link'})
    for result in resultlst:
        title = result.find("strong",{'class':'psil_tit'})
        neirong = result.find("p",{'class':'psil_txt'}).text
        neirong = neirong.replace("\n",'')
        neirong = neirong.replace("\t",'')
        date = result.find("span",{'class':'psil_info'}).text
        date = date.replace("\n",'')
        date = date.replace("\t",'')
        date = date.split("|")[-2]
        date = date.replace(" ",'')

        newslst.append({'title':title, 'date':date,'content':neirong})
    driver.quit()
newsdf = pd.DataFrame(newslst)
newsdf
```

```python
newsdf.to_excel("SBS Crawling_0101.xlsx", index=False)
```
