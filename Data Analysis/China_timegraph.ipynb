```python
import pandas as pd
from datetime import datetime
```

```python
cctv = pd.read_excel("C:/Users/onrik/report/0203_LDA/CCTV 가공.xlsx")
def monthcnt(x):
    if int(x[-2:]) <= 15:
        y = '-01'
    else:
        y = '-15'
    z = str(x)[:7]
    return z+y
cctv['month'] = cctv['date'].apply(lambda x: monthcnt(x))
cctv = cctv[['month','content']]
cctvcnt = []
for m in cctv.month.unique():
    mcnt = len(cctv.loc[cctv['month'] == m])
    cctvcnt.append({'平台':'cctv',"month":str(m), 'count':mcnt})
cctvdf = pd.DataFrame(cctvcnt)
cctvdf.sort_values(by='month', inplace=True)

cctvdf
```

```python
renmin = pd.read_excel("C:/Users/onrik/report/0203_LDA/人民日报 가공2.xlsx")
renmin['month'] = renmin['date'].apply(lambda x: monthcnt(x))
renmin = renmin[['month','content']]
renmincnt = []
for m in renmin.month.unique():
    mcnt = len(renmin.loc[renmin['month'] == m])
    renmincnt.append({'平台':'人民日报',"month":str(m), 'count':mcnt})
renmindf = pd.DataFrame(renmincnt)
renmindf.sort_values(by='month', inplace=True)
renmindf
```

```python
weibo = pd.read_excel("C:/Users/onrik/report/0203_LDA/weibo 가공.xlsx")
weibo.columns = ['date','content']
weibo['month'] = weibo['date'].apply(lambda x: monthcnt(x))
weibo = weibo[['month','content']]
weibocnt = []
for m in weibo.month.unique():
    mcnt = len(weibo.loc[weibo['month'] == m])
    weibocnt.append({'平台':'weibo',"month":str(m), 'count':mcnt})
weibodf = pd.DataFrame(weibocnt)
weibodf.sort_values(by='month', inplace=True)
weibodf
```

```python
chinadf = pd.concat([cctvdf, renmindf, weibodf])
chinadf
```

```python
monthlst = list(chinadf.month.unique())
pingtailst = list(chinadf.平台.unique())
for mon in monthlst:
    for ping in pingtailst:
        if len(chinadf.loc[(chinadf['平台'] == ping)&(chinadf['month']==mon)]) == 0:
            chinadf = chinadf.append({'平台': ping, 'month':mon,'count':0}, ignore_index=True)
chinadf.sort_values(by=['平台','month'],inplace=True)
chinadf
```
